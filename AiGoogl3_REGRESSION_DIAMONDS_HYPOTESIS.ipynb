```python
file_id = "1Lp_Gdc3hRw2FZsHMmFZ0PSJa2bNBh8W6"
url = f"https://drive.google.com/uc?id={file_id}"
gdown.download(url, "data.csv", quiet=False)

diamonds = pd.read_csv("data.csv")
diamonds["log_price"] = np.log(diamonds["price"])
diamonds.head()
```

```python
sns.boxplot(x="color", y="log_price", data=diamonds)
```

```python
model = ols(formula="log_price ~ C(color)", data=diamonds).fit()
```

```python
sm.stats.anova_lm(model,typ=2)
```
